```python
def gradient_desent(x,y):
    m_curr= b_curr=0
    iterations=1000
    n=len(x)
    learning_rate=0.08
    for i in range(iterations):
        yp=m_curr*x+b_curr
        cost=(1/n)*sum([ val**2 for val in (y-yp)])
        md=-(2/n)*sum(x*(y-yp))
        bd=-(2/n)*sum(y-yp)
        
        m_curr=m_curr-learning_rate*md
        print("m {}, b {},cost {}, iterations {}".format(m_curr,b_curr,cost,iterations))
    pass
```
