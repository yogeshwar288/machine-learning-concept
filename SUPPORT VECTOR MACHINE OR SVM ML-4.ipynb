```python
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris_df=load_iris()
```

```python
df=pd.DataFrame(iris_df.data,columns=iris_df.feature_names)
df
```

```python
df.shape
```

```python
df['target']=iris_df.target
```

```python
df.head()
```

```python
iris_df.target
```

```python
df[df.target==1].head()
```

```python
df['flowers_name']=df.target.apply(lambda X: iris_df.target_names[X])
```

```python
df
```

```python
df0=df[:50]
df1=df[50:100]
df2=df[100:150]
```

```python
plt.figure(figsize=(12,6))
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],color='g',marker='*')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],color='b',marker='+')

plt.scatter(df2['sepal length (cm)'],df2['sepal width (cm)'],color='r',marker='*')

plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
```

```python
plt.figure(figsize=(12,6))
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'],color='g',marker='*')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color='b',marker='+')

plt.scatter(df2['petal length (cm)'],df2['petal width (cm)'],color='r',marker='*')

plt.xlabel('patel length (cm)')
plt.ylabel('patel width (cm)')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop(['target','flowers_name'],axis=1)
x
```

```python
x
```

```python
y=df['target']
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.svm import SVC
model=SVC()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

```python
model.predict(X_test)
```
