```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('E:\\csv files\\betulareaa.csv')
```

```python
df
```

```python
dummis=pd.get_dummies(df.town)
dummis
```

```python
dummis=pd.get_dummies(df.town,drop_first=True)
dummis
```

```python
df_dummies=pd.concat([df,dummis],axis=1)
df_dummies
```

```python
df1=df_dummies.drop(['town'],axis=1)
```

```python
df1
```

```python
x=df1.drop('price',axis=1)
x
```

```python
y=df1['price']
y
```

```python
from sklearn import linear_model
```

```python
model=linear_model.LinearRegression()
model.fit(x,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict([[2500,1,0]])
```

```python
import pickle
```

```python
with open('model_pickle','wb') as file:
    pickle.dump(model,file)
```

```python
with open('model_pickle','rb') as file:
    mp=pickle.load(file)
```

```python
mp.predict([[1500,0,0]])
```
