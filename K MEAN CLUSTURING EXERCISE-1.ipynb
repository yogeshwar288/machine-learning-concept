```python
df1=df.drop(['sepal length (cm)','sepal width (cm)','target','flower'],axis=1)
```

```python
df1
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
```

```python
km=KMeans(n_clusters=3)
y_pred=km.fit_predict(df1[['petal length (cm)','petal width (cm)']])
y_pred
```

```python
df1['cluster']=y_pred
df1['flower']=df.flower
```

```python
df1
```

```python
kc=km.cluster_centers_
kc
```

```python
df2=df1[df1.cluster==0]
df3=df1[df1.cluster==1]
df4=df1[df1.cluster==2]
```

```python
sns.scatterplot(x='petal length (cm)',y='petal width (cm)',data=df2)
sns.scatterplot(x='petal length (cm)',y='petal width (cm)',data=df3)
sns.scatterplot(x='petal length (cm)',y='petal width (cm)',data=df4)
sns.scatterplot(x=kc[:,0],y=kc[:,1],data=kc)

```

```python
scaler=MinMaxScaler()
df1['petal length (cm)']=scaler.fit_transform(df[['petal length (cm)']])
df1
```

```python
df1['petal width (cm)']=scaler.fit_transform(df1[['petal width (cm)']])
df1
```

```python
kmc=KMeans(n_clusters=3)
y_predict=kmc.fit_predict(df1[['petal length (cm)','petal width (cm)']])
y_predict
```

```python
df1['cluster']=y_predict
```

```python
kcc=kmc.cluster_centers_
kcc
```

```python
sse=[]
k_rng=range(1,10)
for k in k_rng:
    km=KMeans(n_clusters=k)
    km.fit(df[['petal length (cm)','petal width (cm)']])
    sse.append(km.inertia_)
    plt.plot(sse,marker='*')
```
