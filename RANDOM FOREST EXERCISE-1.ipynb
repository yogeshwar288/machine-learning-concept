```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
df
```

```python
df['target']=iris.target
```

```python
df
```

```python
df['flower']=df.target.apply(lambda X: iris.target_names[X])
```

```python
df
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x='sepal length (cm)',y='sepal width (cm)',data=df,hue='flower',size='flower',sizes=(50,200),style='flower')
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x='petal length (cm)',y='petal width (cm)',data=df,hue='flower',size='flower',sizes=(50,200),style='flower')
```

```python
df
```

```python
x=df.drop(['target','flower'],axis=1)
y=df.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier()
```

RandomForestClassifier(
    n_estimators=100,
    *,
    criterion='gini',
    max_depth=None,
    min_samples_split=2,
    min_samples_leaf=1,
    min_weight_fraction_leaf=0.0,
    max_features='auto',
    max_leaf_nodes=None,
    min_impurity_decrease=0.0,
    min_impurity_split=None,
    bootstrap=True,
    oob_score=False,
    n_jobs=None,
    random_state=None,
    verbose=0,
    warm_start=False,
    class_weight=None,
    ccp_alpha=0.0,
    max_samples=None,
)

```python
model.fit(X_train,y_train)
model.score(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
y_test
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(y_pred,y_test)
cm
```

```python
sns.heatmap(cm,annot=True)
```
