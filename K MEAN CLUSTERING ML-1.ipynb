```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
```

```python
df=pd.read_csv('E:\csv files\income_age.csv')
```

```python
/
df
```

```python
plt.figure(figsize=(15,9))


sns.scatterplot(x='age',y='incomes$',data=df,size='incomes$',sizes=(20,200),hue='age')
```

```python
km=KMeans(n_clusters=3)


y_predicted=km.fit_predict(df[['age','incomes$']])
y_predicted
```

```python
df['cluster']=y_predicted
```

```python
df
```

```python
km.cluster_centers_
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]
```

```python
display(df1,df2,df3)
```

```python
plt.figure(figsize=(15,9))
sns.scatterplot(x='age',y='incomes$',data=df1)
sns.scatterplot(x='age',y='incomes$',data=df2)
sns.scatterplot(x='age',y='incomes$',data=df3)
```

```python
scaler=MinMaxScaler()
scaler.fit(df[['incomes$']])
df['incomes$']=scaler.transform(df[['incomes$']])
scaler.fit(df[['age']])
df['age']=scaler.transform(df[['age']])
df
```

```python
km=KMeans(n_clusters=3)
y_predicted=km.fit_predict(df[['age']],df[['incomes$']])
y_predicted
```

```python
km=KMeans(n_clusters=3)
y_predicted=km.fit_predict(df[['age','incomes$']])
y_predicted
```

```python
df['cluster']=y_predicted
```

```python
df
```

```python
kc=km.cluster_centers_
kc
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]
```

```python
plt.figure(figsize=(15,9))
sns.scatterplot(x='age',y='incomes$',data=df1)
sns.scatterplot(x='age',y='incomes$',data=df2)
sns.scatterplot(x='age',y='incomes$',data=df3)
sns.scatterplot(x=kc[:,0],y=kc[:,1],data=kc)
```

```python
sse=[]
k_rng=range(1,10)
for k in k_rng:
    km=KMeans(n_clusters=k)
    km.fit(df[['age','incomes$']])
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.plot(k_rng,sse,marker='*')
```
